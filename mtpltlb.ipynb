```python
import pandas as pd
df=pd.read_csv('Used_Bikes.csv')
df.head(3)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['owner'].value_counts()
```

```python
ownership=list(df['owner'].value_counts().keys())
ownership
```

```python
no_of_bikes=list(df['owner'].value_counts().values)
no_of_bikes
```

```python
plt.figure(figsize=(10,6))
plt.bar(ownership,no_of_bikes)
plt.xlabel("Ownership")
plt.ylabel("No. of Bikes")
plt.title("Variation of no of bikes acc to Owner")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(ownership,no_of_bikes,color='green',ec='k',lw=4,ls='dotted')       #ec=edge color and K stands for BLACK , by default edge is of same color as the bar
#lw=line width of the edges
#ls=line style of the edges => DASHED,DOTTED
plt.xlabel("Ownership")
plt.ylabel("No. of Bikes")
plt.title("Variation of no of bikes acc to Owner")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
plt.pie(country_pop,labels=country_name)        #PIE PLOT=same as bar plot(need one categorical and one numerical col)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(country_pop,labels=country_name,autopct='%2.3f%%',wedgeprops={'ec':"black",'ls':'dashed','lw':3})      #autopct=automatically calculate percentage      #first 2 means for int values show 2 no. before decimal and second 2 with f means for float values show 2 no. after decimal alongwith a % sign      #it always starts and ends with a % sign
#wedgeprops=edge properties
plt.show()
```

```python
print(country_name)
print(country_pop)
explosion=[0,0,0,0]

for i in range(0,len(country_name)):
    if country_name[i]=="India":
        explosion[i]=0.2     
     
plt.figure(figsize=(5,5))
plt.pie(country_pop,labels=country_name,autopct='%2.3f%%',wedgeprops={'ec':"white",'lw':3},explode=explosion,shadow=True)       #explode is generally used whenever we want to highlight some specific data
#shadow=True adds a shadow
plt.savefig("pieplot_image")
plt.show()
```

```python
#SARABJEET QUESTION-3rd AUG
import pandas as pd
df=pd.read_csv('Used_Bikes.csv')
mindf = df.groupby('brand')['price'].min()
plt.figure(figsize=(10, 6))     #dimensions of plot     #width,height in inches
plt.plot(mindf.index, mindf.values, marker='o',ls='dotted')  
plt.xlabel('Brand')
plt.ylabel('Minimum Price')
plt.title('Minimum Price by Vehicle Brand')
plt.xticks(rotation=45)  
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#INTERNSHALA - 4th AUGUST
day=[1,2,3,4,5]
temp=[39,41,27,22,17]
#PLOT THE DATA
plt.plot(day,temp,marker='o',linestyle='-',color='b',label='Temperature')       #b stands for blue
#ADDING LABELS AND TITLE
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Temperature Variation")
#DISPLAY THE PLOT
plt.legend()        #gives reference or scale for diff colours we are going to use
plt.grid(True)      #used to add grids or sqaure boxes in the background
plt.show()
```
